```python
original_data = pd.read_csv('../dataset/mohler2_cleaned.csv')
original_data = original_data.drop(labels='Unnamed: 0', axis=1)
original_data = original_data.rename(columns={'question_number':'question_id','question_text':'question','answer_model':'ref_answer','answer_student':'student_answer','score_avg':'grade'})
```

```python
df = original_data
#converting to lower case
df['ref_answer'] = df['ref_answer'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df['student_answer'] = df['student_answer'].apply(lambda x: " ".join(x.lower() for x in x.split()))

#punctuation removal
df['ref_answer'] = df['ref_answer'].str.replace('[^\w\s]','')
df['student_answer'] = df['student_answer'].str.replace('[^\w\s]','')

#stop word removal
stop = stopwords.words('english')
df['ref_answer'] = df['ref_answer'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
df['student_answer'] = df['student_answer'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))


# #lemmatisation
df['ref_answer'] = df['ref_answer'].apply(lambda x: " ".join([Word(word).lemmatize() for word in word_tokenize(x)]))
df['student_answer'] = df['student_answer'].apply(lambda x: " ".join([Word(word).lemmatize() for word in word_tokenize(x)]))

# #stemming
# st = PorterStemmer()
# df['ref_answer'] = df['ref_answer'].apply(lambda x: " ".join([st.stem(word) for word in word_tokenize(x)]))
# df['student_answer'] = df['student_answer'].apply(lambda x: " ".join([st.stem(word) for word in word_tokenize(x)]))
df
```
