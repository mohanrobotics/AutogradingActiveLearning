```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# total = np.vstack(( [9,9,9,3,9,9],[2,6,7,4,8,5], [4,6,6,4,7,3],[1,4,4,4,6,8]))
total = np.vstack(( [1,1,1,1,1,1],[4,4,4,0,4,4], [2,2,2,2,2,2], [3,3,3,3,3,3]))

# competitors = ["Logistic - Margin","NaiveBayes - Margin",\
#                "Random forest - Least confident","SVC Linear - Margin",\
#                "Committee - Max disgreement"]
tot_dict = {i:[] for i in range(total.shape[0])}

print(total)
print(np.argsort(total[:,0]))
print(np.argsort(total[:,1]))
print(np.argsort(total[:,2]))
print(np.argsort(total[:,3]))
print(np.argsort(total[:,4]))
print(np.argsort(total[:,5]))





freq = 1

for i in range(0,6):
    ranks = np.argsort(total[:,int(i)])
    for j in range(total.shape[0]):
        tot_dict[j].append(ranks[j])
        
print(tot_dict)
        
labels = ["1","4", "2","3"]

fig = plt.figure(figsize=(18,6))
for i in range(total.shape[0]):
    plt.plot(range(0,len(tot_dict[i])), tot_dict[i], "o-"\
             , label=labels[i],markersize=10, linewidth=2)
plt.legend()    
plt.show()
```
