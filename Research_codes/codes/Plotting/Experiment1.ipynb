```python
import matplotlib.pyplot as plt
import matplotlib as m
import numpy as np
import pickle as pkl
```

```python
#plotting
dict_accuracy_al = pkl.load( open("../../results/dict_accuracy_al_mohler_"+str(80)+".pkl","rb"))
dict_accuracy_sl_min = pkl.load( open("../../results/dict_accuracy_min_mohler.pkl","rb"))
dict_accuracy_sl_max = pkl.load( open("../../results/dict_accuracy_max_mohler.pkl","rb"))
dict_accuracy_sl_avg = pkl.load( open("../../results/dict_accuracy_avg_mohler.pkl","rb"))
dict_accuracy_al_random = pkl.load( open("../../results/random_sampling/Multi_class/dict_accuracy_random_mohler_"+str(80)+".pkl","rb"))

colors = [(148, 103, 189),(31, 119, 180),(197, 176, 213), (196, 156, 148),(227, 119, 194)]

for i in range(len(colors)):    
    r, g, b = colors[i]    
    colors[i] = (r / 255., g / 255., b / 255.) 

## plotting
fig = plt.figure(figsize=(14,16))
model_title = ["Logistic regression","Naive Bayes","Random Forest","SVC-linear","SVC"]
query_strategy = ["least confident sampling","margin sampling","entropy sampling"]
for i in range(0,5):
    ax = fig.add_subplot(3,2,i+1)
    for j in range(len(query_strategy)):
        ax.plot(dict_accuracy_al[i][j],color = colors[j],label = "Active learner {}".format(query_strategy[j]))
    ax.plot(dict_accuracy_al_random[i][0], label = "Random learner",linestyle=(0, (5, 5)) ,color="tab:brown", alpha=0.5)
#     ax.legend(loc="lower right")
    ax.set_xlabel("Queried answers for grading (%)")
    ax.set_ylabel("Accuracy %")
    ax.set_title(model_title[i])
    ax.grid(color='g', linestyle='-', linewidth=0.2)
    ax.set_ylim(0,1)
    ax.set_xticks([0,int(len(dict_accuracy_al[0][0])/4),int(len(dict_accuracy_al[0][0])/2),\
           int(len(dict_accuracy_al[0][0])/1.33),int(len(dict_accuracy_al[0][0]))])
    ax.set_xticklabels(['0','20%','40%','60%','80%'])
plt.legend(bbox_to_anchor=(2.10, 0.65),prop = {'size':14})
fig.savefig("./images/task1_accuracy_uncertainty")
plt.show()


# ================== Commitee based learner ====================

#plotting committee
dict_accuracy_al_com = pkl.load( open("../../results/dict_accuracy_al_com_mohler_"+str(80)+".pkl","rb"))
dict_accuracy_al_random = pkl.load( open("../../results/random_sampling/Multi_class/dict_accuracy_random_com_mohler_"+str(80)+".pkl","rb"))


## plotting
fig = plt.figure(figsize=(12,8))
query_strategy = ["vote entropy sampling","max disagreement sampling","consensus entropy sampling"]
for j in range(len(query_strategy)):
    plt.plot(dict_accuracy_al_com[j][0],color = colors[j],label = "Active learner {}".format(query_strategy[j]))
plt.plot(dict_accuracy_al_random[0][0], label = "Random committee learner",\
        linestyle=(0, (5, 5)) ,color="tab:brown",alpha = 0.5 )
plt.legend(loc="lower right")
plt.xlabel("Queried answers for grading (%)")
plt.ylabel("Accuracy %")
plt.title("Committee")
plt.grid(color='g', linestyle='-', linewidth=0.2)
plt.ylim(0,1)

plt.xticks([0,int(len(dict_accuracy_al[0][0])/4),int(len(dict_accuracy_al[0][0])/2),\
       int(len(dict_accuracy_al[0][0])/1.33),int(len(dict_accuracy_al[0][0]))],\
           ['0','20%','40%','60%','80%'])
# plt.xticklabels(['0','20%','40%','60%','80%'])
    
plt.show()
fig.savefig("./images/task1_accuracy_com")
```

```python
total = np.vstack((dict_accuracy_al[0][1],dict_accuracy_al[1][1] ,\
                  dict_accuracy_al[2][0], dict_accuracy_al[3][1],\
                  dict_accuracy_al_com[1][0]))

competitors = ["Logistic - Margin","NaiveBayes - Margin",\
               "Random forest - Least confident", "SVC Linear - Margin",\
               "Committee - Max disgreement"]
tot_dict = {i:[] for i in range(total.shape[0])}

freq = 75

for i in np.linspace(0,total.shape[1]-1,freq):
    ranks = np.argsort(total[:,int(i)])
    ranks = np.argsort(ranks)
    for j in range(total.shape[0]):
        tot_dict[j].append(ranks[j])

fig = plt.figure(figsize=(18,6))
for i in range(total.shape[0]):
    plt.plot(range(0,len(tot_dict[i])), tot_dict[i], "o-", \
             label = competitors[i], markersize=10, linewidth=2)

plt.xticks([0,int(freq/4),int(freq/2),\
           int(freq/1.33),int(freq)],\
           ['0','20%','40%','60%','80%'])
labels = [i for i in range(total.shape[0])]
label_names = reversed([str(i+1) for i in range(total.shape[0])])
plt.yticks(labels,label_names)
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3,\
           ncol=3, mode="expand", borderaxespad=0.)
plt.xlabel("Queried answers for grading (%)")
plt.ylabel("Rank according to accuracy")
plt.tight_layout()
fig.savefig("./images/task1_rank")
plt.show()
```
